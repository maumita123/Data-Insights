```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
ds = pd.read_csv("C:\\Users\\C H DAS\\Downloads\\SampleSuperstore.csv")
fig,axes = plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Total profit Vs Sales")
sns.barplot(data=ds.groupby('Sub-Category')['Sales','Profit'].agg(sum),x = 'Sales',y='Profit',ax=axes[1])
ds.groupby('Sub-Category')['Sales','Profit'].agg(sum).plot(kind='bar',ax = axes[0])
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
ds = pd.read_csv("C:\\Users\\C H DAS\\Downloads\\SampleSuperstore.csv")
fig,axes = plt.subplots(1,2,figsize=(14,5))
ds.groupby('Sub-Category')['Profit','Quantity'].agg(sum).plot(kind='bar',ax=axes[0]).set_title('Quantity and Profit Relation based')
ds.groupby('Sub-Category')['Profit','Discount'].agg(sum).plot(kind='bar',ax=axes[1]).set_title('Discount and Profit Relation based')
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes = plt.subplots(2,2,figsize=(14,5))
fig.suptitle("Distribution of plot", fontsize=16)
sns.displot(ds['Sales'],ax=axes[0,0])
sns.displot(ds['Profit'],ax=axes[0,1])
sns.displot(ds['Quantity'],ax=axes[1,0])
sns.displot(ds['Discount'],ax=axes[1,1])
plt.show()
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,7))
sns.countplot(ds['Quantity'],hue=ds['Region'])
plt.show()
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,8))
fig.suptitle("Sales with different shipping modes and segments")
sns.barplot(ds['Ship Mode'],ds['Sales'],ax=axes[0,0])
sns.lineplot(ds['Ship Mode'],ds['Sales'],ax=axes[0,1])
sns.barplot(ds['Segment'],ds['Sales'],ax=axes[1,0])
sns.lineplot(ds['Segment'],ds['Sales'],ax=axes[1,1])
plt.show()
```

### Important Observations

#### Features like Profit and Discounts are highly related.
#### Over Less quantity of products also the sales are high.
#### Maximum quantity  of products  demand is in range 2-4.
#### Mode of Shipping doesn't affect much to sales.
#### The Home Office Segment provides highest Sales followed By Coporate and Consumer.
